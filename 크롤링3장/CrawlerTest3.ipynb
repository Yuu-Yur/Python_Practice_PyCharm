```python
import urllib.request
import bs4
```

```python
nateUrl = "https://www.nate.com"
# url 열기
htmlObject = urllib.request.urlopen(nateUrl)
# 읽기 읽으면 이해하기어려움
html = htmlObject.read()
html
```

```python
# beautifulsoup 를 이용해서 parser 사용해 이해하기 쉽게 파싱(변환)
bsObj = bs4.BeautifulSoup(html, "html.parser")
# parser 은 html.parser , lxml, lxml-xml(xml), html5lib
# 뒤의 세개는 추가설치해야하지만 속도가 더빠르고 html5lib는 브라우저와 동일한 방식으로 파싱 지원
bsObj
```
