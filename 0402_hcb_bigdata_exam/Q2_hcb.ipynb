```python
# 2.공공데이터 Open API를 활용한, 출입국 데이터 수집해서 json 파일에 저장하기.
import urllib.request
import urllib.parse
import json
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
# serviceKey = os.getenv('PUBLIC_DATA_ENCODING_KEY')
# serviceKey = os.getenv('PUBLIC_DATA_DECODING_KEY')
serviceKey = os.getenv('TEMP_KEY')
```

```python
def call_api(yyyymm):
    url = "http://openapi.tour.go.kr/openapi/service/EdrcntTourismStatsService/getEdrcntTourismStatsList?_type=json"
    parameters = "&serviceKey=%s&YM=%s" % (serviceKey, yyyymm)
    url += parameters
    request = urllib.request.Request(url)
    try:
        response = urllib.request.urlopen(request)
        data = json.loads(response.read().decode("utf-8"))
        if response.getcode() == 200:
            return data if len(data)>0 else None
    except Exception as e:
        print(e)
```

```python
def getAlldata(thisYear):
    dict_list = []
    isDataEnd = 0
    for year in range(thisYear, 0,-1):
        for month in range(1, 13):
            if(isDataEnd == 1): break    #수집기간이 남았어도 데이터 끝이면 작업 중지.

            yyyymm = "{0}{1:0>2}".format(str(year), str(month))
            jsonData = call_api(yyyymm)

            if (jsonData['response']['header']['resultMsg'] == 'OK'):
                # 아직 입력된 범위까지 수집하지 않았지만, 더이상 제공되는 데이터가 없는 경우 -------------------
                if jsonData['response']['body']['totalCount'] == 0:
                    isDataEnd = 1 #데이터 끝 flag 설정
                    if (month-1)==0:  #마지막 데이터가 12월인데, 현재 month가 다음해 1월인 경우
                        year = year-1
                        month = 13

                    dataEND = "{0}{1:0>2}".format(str(year), str(month-1))
                    jsonData = call_api(dataEND) #출력용으로 마지막 데이터 다시 호출
                    break

                #반환할 데이터 항목을 추출하여 저장
                for item in jsonData['response']['body']['items']['item']:
                    natName = item['natKorNm']
                    natName = natName.replace(' ', '')
                    num = item['num']
                    ed = item['ed']
                    dict_list.append({'nat_name': natName, 'yyyymm': yyyymm, 'visit_cnt': num})
                print(f"{yyyymm}완료")
    return (dict_list)
```

```python
def main():
    yyyy = 2024
    result_dict_list = getAlldata(2024)
    with open('Q2_hcb.json' , 'w', encoding='utf8') as f:
        jsonFile = json.dumps(result_dict_list, indent = 4, sort_keys = True, ensure_ascii = False)
        f.write(jsonFile)

if __name__ == '__main__':
    main()
```
